# IMDB Top 250 Web Scraping

```python
from bs4 import BeautifulSoup
import requests
import time
import datetime
import smtplib
import numpy as np
import pandas as pd
```

```python
master_url = 'https://www.imdb.com/chart/top/'
url_imdb = 'https://www.imdb.com/title/tt0111161/?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=e31d89dd-322d-4646-8962-327b42fe94b1&pf_rd_r=49X3PDYPNY42GGJ1Z43N&pf_rd_s=center-1&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_tt_1'
#user agent:
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36",}

page0 = requests.get(master_url,headers=headers)
page2 = requests.get(url_imdb,headers=headers)
```

```python
master_url = 'https://www.imdb.com/chart/top/'
#pulling doctype html

soup0 = BeautifulSoup(page0.content,'html.parser')
soup0 = BeautifulSoup(soup0.prettify(),'html.parser')
```

```python
movies_links = []
a_href= soup0.find('tbody',class_= 'lister-list').find_all('tr')
for i in range(len(a_href)):
  
 href=a_href[i].find('a')['href']
 href= 'https://www.imdb.com'+ href
 movies_links.append(href)
 
```

```python
data_headers = ['Title', 'Rating','Oscars','Total Nominations and Awards','Director']
movie_title=[]
movie_rate=[]
oscars_list=[]
directors_list=[]
'ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link'
nomination_wins_total=[]
for i in movies_links:
    sub_page= requests.get(i,headers=headers)
    sub_soup = BeautifulSoup(sub_page.content,'html.parser')
    sub_soup = BeautifulSoup(sub_soup.prettify(),'html.parser')
    
    title = sub_soup.find('title').get_text()
    title=title.strip()
    title=title[0:-len( ' - imdb')]

     
    rating = sub_soup.find('div',class_= 'AggregateRatingButton__ContentWrap-sc-1ll29m0-0 hmJkIS').find('span').get_text()
    oscars = sub_soup.find('ul',class_= 'ipc-metadata-list ipc-metadata-list--dividers-none Awards__Content-sc-1qdt65t-2 pWPhK ipc-metadata-list--base').find('a').get_text()
    nom_wins =sub_soup.find('span',class_= 'ipc-metadata-list-item__list-content-item').get_text()
    director =sub_soup.find('a',class_= 'ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link').get_text()
  
    
  
    oscars=oscars.strip()
    oscars_list.append(oscars)   
    nom_wins=nom_wins.strip()
    director=director.strip()

    rating= rating.strip()
    movie_title.append(title)
    movie_rate.append(rating)
    nomination_wins_total.append(nom_wins)
    directors_list.append(director)
    
    myvalues =[movie_title,movie_rate,oscars_list,nomination_wins_total,directors_list]
   

data_dict =dict(zip(data_headers,myvalues))  
```
